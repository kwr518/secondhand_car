```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained('paust/pko-t5-large')
model = T5ForConditionalGeneration.from_pretrained('paust/pko-t5-large')

# 입력 데이터 (자동차 정보)
car_data = {
    "모델": "쏘나타",
    "연식": "2018",
    "주행거리": "50,000km",
    "사고이력": "없음",
    "실제판매가격": "1,200만 원",
    "적정예측가격": "1,300만 원"
}

input_text = f"""
이 차량은 {car_data['연식']}년식 {car_data['모델']}입니다.
주행거리는 {car_data['주행거리']}이며, 사고 이력은 {car_data['사고이력']}입니다.
적정 예측 가격은 {car_data['적정예측가격']}이고, 실제 판매 가격은 {car_data['실제판매가격']}입니다.
"""

# T5 작업 태스크 설정
task_prefix = "생성: "
input_ids = tokenizer(task_prefix + input_text, return_tensors="pt").input_ids

# 텍스트 생성
output_ids = model.generate(
    input_ids,
    max_length=150,
    num_beams=5,
    early_stopping=True
)

# 생성된 텍스트 디코딩
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("생성된 문장:")
print(output_text)

```
